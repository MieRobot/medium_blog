```
# Importing required libraries
import cv2 
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import numpy as np
```

```
## reading images 
img = cv2.imread('sudoku_raw.png',0) 
img1=cv2.imread('img2.png',0)

## Displaying read images using plt.imshow
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.title("Image 1")

plt.subplot(1,2,2)
plt.imshow(img1,cmap='gray')
plt.title("Image 2")

plt.setp(plt.gcf().get_axes(),xticks=[],yticks=[])
```

```
print('Image 1',img.shape)
plt.imshow(img,cmap='gray')
```

```
print('Image 2',img1.shape)
plt.imshow(img1,cmap='gray')
```

```
## Function that takes in image on which kernel operation should be performed 
#and kernel values corresponding to the type of operation that is to be performed. 

def edge_detect(img,kernel):
"""
  This function displays transformed output of image based on kernel supplied.

  Parameters:
  img(ndarray) : numpy array representing grayscale image
  kernel(ndarray) : 3*3 numpy array represnting kernel operation

  Returns:
  None
"""

  # saving output of convolution
  res=convolve2d(img,kernel)

  # show original and transformed images
  plt.subplot(1,2,1)
  plt.imshow(img,cmap="gray")
  plt.title("original")

  plt.subplot(1,2,2)
  plt.imshow(res,cmap='gray')
  plt.title("transformed")
  
  # removing x and y axis ticks
  plt.setp(plt.gcf().get_axes(),xticks=[],yticks=[])
```

```
## Using Horizontal Kernel
h_kernel=np.array([[-1,-2,-1],
                   [0,0,0],
                   [1,2,1]])
edge_detect(img,h_kernel)
```

```
## Using vertical kernel
v_kernel=np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])
edge_detect(img,v_kernel)
```

```
## Extracting diagonals 
diag_kernel=np.array([[0,1,2],
                      [-1,0,1],
                      [-2,-1,0]])
edge_detect(img1,diag_kernel)
```
